# How to pull the full dataset

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
allData = pd.read_csv('https://media.githubusercontent.com/media/DesiPilla/MachineLearning/master/Project/fullStockData.csv').drop(columns = ['Unnamed: 0'])
allData['Date'] = pd.to_datetime(allData['Date'])
```

```
allData.head()
```

# Examples of how to query subsets

**Example of a subset for a single stock**

```
ticker = 'AMZN'
stockData = allData[allData['Ticker'] == ticker]
print(stockData.shape)
stockData.head()
```

**The 11 GICS Sectors**

**Example of a subset for a single sector**

```
print('The 11 sectors (plus unknown) are:')
for i in allData['Sector'].unique():
  print('\t',i)
```

```
sector = 'Information Technology'
sectorData = allData[(allData['Sector'] == 'Information Technology')]
print(sectorData.shape)
sectorData.head()
```

**Example of a subset between specific dates**

```
start = pd.to_datetime('1-1-2000')
end = pd.to_datetime('4-1-2000')
q1Data = allData[(start <= allData['Date']) & (allData['Date'] < end)]
q1Data
```
